```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy.io import loadmat

import stfft
import general
import eegFilter

from scipy.io import wavfile
```

```python
filename = "archive/XC377470-European-Robin-Erithacus-rubecula.wav"
fs, birdsound = wavfile.read(filename)
normdata = birdsound[:,0]/np.max(birdsound[:,0])

pxx = stfft.dwt_tf(np.reshape(normdata[0:10*fs], (1,10*fs)), fs, np.logspace(0,np.log10(15000),50), (0.5,1))

plt.figure(figsize=(15,5))
# plt.imshow(pxx[:,:], aspect='auto', cmap=plt.get_cmap('gray_r'))
plt.contourf(np.linspace(0,10,10*fs),np.logspace(0,np.log10(15000),50), pxx, cmap=plt.get_cmap('gray_r'))
plt.yticks(np.linspace(0,15000,6), ['0','3k', '6k','9k','12k','15k'])
# plt.ylim([0,100])
plt.show()
```

```python
filename = "archive/Woodcreeper.wav"
fs, Woodcreeper = wavfile.read(filename)

plt.plot(Woodcreeper)
plt.show()

pxx = stfft.dwt_tf(np.reshape(Woodcreeper, (1,np.size(Woodcreeper))), 
                   fs, np.logspace(0,np.log10(15000),50), (1/fs,1000/fs))

plt.figure(figsize=(15,5))
# plt.imshow(pxx[:,:], aspect='auto')
plt.contourf(np.linspace(0,np.size(Woodcreeper)/fs,np.size(Woodcreeper)),
            np.logspace(0,np.log10(15000),50), pxx, 30)
# plt.yticks(np.linspace(0,15000,6), ['0','3k', '6k','9k','12k','15k'])
# # plt.ylim([0,100])
plt.show()
```
