```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_list = range(2,501)
trialN = 1000
result = np.zeros((len(N_list), trialN))
for Nidx, eachN in enumerate(N_list):
    for trial_idx in range(trialN):
        ind = np.exp(1j*np.random.rand(eachN)*2*np.pi)
        itpc = np.abs(np.sum(ind) / eachN)
        result[Nidx, trial_idx] = itpc

itpc_mu = np.mean(result,1)
itpc_sig = np.std(result,1)

plt.figure(figsize=(18,6))
plt.plot(N_list, itpc_mu)
# plt.plot(N_list, itpc_mu+itpc_sig*2, linestyle='--')
plt.fill_between(N_list, itpc_mu-itpc_sig, itpc_mu+itpc_sig, alpha=0.3)
plt.ylim([0,1])
plt.xlim([1, N_list[-1]])
plt.ylabel("ITPC")
plt.xlabel("N")
plt.show()
```
